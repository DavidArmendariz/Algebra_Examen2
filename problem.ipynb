```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_matrix(n):
    A = np.zeros((n, n))
    for i in range(n):
        A[:, i] = np.arange(1, n +1) ** (i + 1)
    return A
```

```python
def condition_number(A):
    return np.linalg.cond(A)
```

```python
x = []
y = []
for i in range(1, 101):
    A = generate_matrix(i)
    x.append(i)
    y.append(condition_number(A))
```

```python
plt.plot(x, y)
plt.title("Número de condición vs n")
plt.xlabel("n")
plt.ylabel("Número de condición")
plt.show()
```
